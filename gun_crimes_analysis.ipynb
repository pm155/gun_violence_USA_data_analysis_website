```python
def map(df):
    from geopy.geocoders import Nominatim
    geolocator = Nominatim(user_agent="location_details")
    df['Address'] = df['city'] + ',' + df['state']
    df = df[df['total victims'] > 11]
    def func(obj):
        from geopy.geocoders import Nominatim
        loc = Nominatim(user_agent="GetLoc")
        getLoc = loc.geocode(obj)

        a = getLoc.latitude
        b = getLoc.longitude
        return a, b

    df['co-ordinate'] = df['Address'].apply(func)
    df['lat'] = [g[0] for g in df['co-ordinate']]
    df['long'] = [g[1] for g in df['co-ordinate']]
    return df
```

```python
df=map(df)
```

```python
def top_10(df,col):
    data = df[[col,'n_killed']].nlargest(10, 'n_killed')[col].to_frame().reset_index().rename(columns={"index": "n_killed"})
    return data
```
